```python
import seaborn as sns
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = np.random.normal(loc=50, scale=10, size=1000)

# Plot with fitted normal distribution
sns.displot(data, kde=True, fit=stats.norm)
plt.title("Distribution with Fitted Normal Curve")
plt.show()

# Get fitted parameters for normal distribution
mean, std = stats.norm.fit(data)
print(f"Fitted Normal Distribution: Mean = {mean:.2f}, Standard Deviation = {std:.2f}")

# Plot with fitted gamma distribution
sns.displot(data, kde=True, fit=stats.gamma)
plt.title("Distribution with Fitted Gamma Curve")
plt.show()

# Get fitted parameters for gamma distribution
shape, loc, scale = stats.gamma.fit(data)
print(f"Fitted Gamma Distribution: Shape = {shape:.2f}, Location = {loc:.2f}, Scale = {scale:.2f}")
```
